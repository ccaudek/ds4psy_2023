### Jupyter Notebook

[Jupyter Notebook](https://www.dataquest.io/blog/jupyter-notebook-tutorial/) è un'applicazione web open-source utilizzata principalmente per la scrittura di codice e la creazione di documenti interattivi. Ciò significa che i Jupyter Notebook consentono di creare e condividere documenti contenenti codice, testo descrittivo, grafici e altri elementi multimediali, rendendoli un ottimo strumento per la documentazione e la condivisione dei processi di analisi dei dati in modo trasparente e riproducibile.

In un Jupyter Notebook, il codice viene eseguito in celle, che possono essere eseguite in modo interattivo e in qualsiasi ordine. Ciò significa che gli utenti possono esplorare i dati e fare analisi dei dati in modo iterativo, aiutando a capire e ad affinare le proprie analisi. Inoltre, i Jupyter Notebook consentono di integrare facilmente codice proveniente da molte lingue di programmazione diverse, come Python, R, Julia e altri.

Il Jupyter Notebook è stato originariamente sviluppato come parte del progetto IPython e prende il nome dalle tre principali lingue di programmazione che supporta: Julia, Python e R. Tuttavia, il Jupyter Notebook può essere utilizzato con molte altre lingue di programmazione.

In particolare, il Jupyter Notebook supporta due tipi di celle: celle di testo e celle di codice.

- Le celle di testo permettono di scrivere testo formattato utilizzando la sintassi Markdown. Questo consente agli autori di scrivere del testo descrittivo, includere immagini, formule in formato $\LaTeX$, creare tabelle e altro ancora. In questo modo, le celle di testo permettono di documentare il processo di analisi dei dati in modo chiaro e facile da comprendere.
- Le celle di codice, d'altra parte, consentono di scrivere e eseguire codice Python. Il codice può essere eseguito cliccando sul triangolo presente a sinistra di ogni cella. In diverse celle possono essere aggiunte istruzioni diverse e le celle possono essere eseguite sequenzialmente. Una funzione definita in una cella precedente può essere utilizzata successivamente solo se la cella precedente è stata eseguita.

I Jupyter Notebook possono essere eseguiti sia localmente, sul proprio computer, che su un server remoto, come Colab di Google. In questo modo, gli utenti possono accedere ai propri notebook da qualsiasi dispositivo connesso a Internet e condividere facilmente il proprio lavoro con altri utenti. Risultano già installate su Colab numerose librerie Python in ogni runtime. Capita però di avere bisogno di installare librerie aggiuntive su Colab. Le librerie possono essere scaricate usando i comandi bash con `!`. Questo dice a Colab che questa cella del notebook non è un codice Python ma uno script della riga di comando. Per esempio, possiamo installare le librerie `pymc`, `bambi` e `arviz` nel modo seguente.

````
!pip install bambi
!pip install pymc
!pip install arviz
````

#### Kernel

In un Jupyter Notebook, il kernel è il componente software che esegue il codice Python all'interno delle celle di codice. Quando un utente esegue una cella di codice, il contenuto della cella viene inviato al kernel per essere eseguito. Il kernel è in grado di mantenere lo stato delle variabili e delle funzioni Python tra l'esecuzione di una cella e l'altra, in modo che le variabili e le funzioni definite in una cella possano essere utilizzate in celle successive. Questo consente agli utenti di eseguire codice in modo interattivo e di esplorare i dati in modo iterativo. Il Jupyter Notebook supporta diversi kernel, tra cui quello di Python, ma anche di altri linguaggi di programmazione come R, Julia e Matlab. In questo modo, gli utenti possono utilizzare il Jupyter Notebook anche per l'analisi dei dati in altri linguaggi di programmazione.

In sintesi, il kernel è il componente centrale di un Jupyter Notebook che esegue il codice Python all'interno delle celle di codice e mantiene lo stato delle variabili e delle funzioni tra l'esecuzione di una cella e l'altra.

#### Markdown

Markdown è un linguaggio di markup leggero, utilizzato per formattare il testo in modo semplice ed efficiente. Il linguaggio Markdown permette di creare documenti con formattazione di testo senza dover utilizzare codice HTML complesso, ed è molto popolare in ambienti di scrittura tecnica, inclusi i notebook Jupyter.

Il linguaggio Markdown è costituito da una serie di simboli e convenzioni di formattazione che permettono di creare titoli, elenchi, tabelle, citazioni, testo in grassetto e in corsivo e altro ancora. Ecco alcuni esempi di come è possibile formattare il testo in Markdown:

- Per creare un titolo, basta utilizzare il simbolo "#" seguito dal testo del titolo. Ad esempio, `# Titolo del documento` creerà un titolo di primo livello.
- Per creare una lista con punti, basta utilizzare il simbolo "-" seguito dal testo dell'elemento della lista. Ad esempio, `- Elemento 1` creerà un elemento di lista con un punto.
- Per creare un link ipertestuale, basta utilizzare il simbolo "[" seguito dal testo del link e dal simbolo "]" seguito dall'URL del link. Ad esempio, `[Link a Google](https://www.google.com/)` creerà un link con il testo "Link a Google" che porta alla homepage di Google.
- Per creare testo in grassetto, basta utilizzare il simbolo "**" prima e dopo il testo da formattare. Ad esempio, `**Testo in grassetto**` creerà il testo in grassetto.
- Per creare testo in corsivo, basta utilizzare il simbolo "*" prima e dopo il testo da formattare. Ad esempio, `*Testo in corsivo*` creerà il testo in corsivo.
  
In sintesi, il linguaggio Markdown è un modo semplice ed efficace per formattare il testo senza dover utilizzare codice HTML complesso. Il Markdown è molto popolare in ambienti di scrittura tecnica come i notebook Jupyter, poiché consente di creare documenti facilmente leggibili e con una formattazione coerente e pulita.

#### Command magic

I "command magic" sono una serie di comandi speciali che possono essere utilizzati all'interno di Jupyter Notebook e JupyterLab per eseguire diverse azioni. Ecco una breve spiegazione dei command magic più comuni:

- `%run`: questo comando consente di eseguire un file Python come programma.
- `%timeit`: questo comando esegue una riga di codice più volte e ne calcola il tempo medio di esecuzione.
- `%matplotlib`: questo comando abilita l'integrazione di Matplotlib per la visualizzazione dei grafici all'interno del notebook.
- `%load`: questo comando consente di caricare il codice di un file esterno all'interno di una cella del notebook.
- `%reset`: questo comando rimuove tutte le variabili presenti nella sessione corrente del notebook.
- `%pwd`: questo comando restituisce il percorso della directory corrente.
- `%cd`: questo comando cambia la directory di lavoro corrente.
- `%%time`: questo comando calcola il tempo di esecuzione di un'intera cella di codice.

Ci sono molti altri comandi magic disponibili in Jupyter Notebook e JupyterLab, e potete visualizzarli tutti digitando `%lsmagic` all'interno di una cella del notebook.

### Anaconda

Il modo più semplice per installare Python è quello di usare [Anaconda](https://www.anaconda.com/). Per prima cosa, dunque, è necessario procedere con l'installazione di Anaconda sul vostro computer.


### L'ambiente virtuale

Dopo aver completato l'installazione di Python, è necessario impostare un ambiente virtuale, chiamato "virtual environment", all'interno del quale è possibile installare l'interprete e le librerie necessarie. Il virtual environment è un "contenitore" separato dal resto del sistema operativo, il che significa che le librerie installate all'interno del virtual environment non interferiranno con altre installazioni globali del sistema.

La ragione per cui si preferisce utilizzare un virtual environment invece di un'installazione globale è legata alla rapida evoluzione del mondo Python. Spesso, infatti, ci sono differenze sostanziali anche tra le release minori dell'interprete, il che può rendere incompatibili le librerie e, di conseguenza, i programmi scritti per le diverse versioni. Avere un ambiente deterministico che specifica la versione dell'interprete e le versioni delle librerie rappresenta una "garanzia" di funzionamento dei programmi. Basterà, infatti, replicare esattamente la configurazione dell'ambiente virtuale per garantire il corretto funzionamento del programma.

Di seguito descrivo, passo dopo passo, la procedura per gestire gli ambienti virtuali utilizzando il programma `conda`. Sia `conda` che `pip` sono inclusi in Anaconda e saranno disponibili sul computer una volta che Anaconda è stato installato. Per eseguire queste istruzioni, è necessario utilizzare il *terminale* (*Prompt dei comandi* su Windows).

Per prima cosa, è necessario uscire dall'ambiente virtuale corrente:

```bash
conda deactivate
```

Attenzione: questo è un passaggio **necessario** che non si deve dimenticare. Se ottenete un messaggio che vi avverte che non si può deattivare alcun ambiente virtuale, non preoccupatevi: il messaggio vi dice semplicemente che l'istruzione precedente non era necessaria.

A questo punto possiamo creare un nuovo ambiente virtuale che è possibile chiamare, ad esempio, `pymc_511` (il nome è arbitrario). 

```bash
conda create -c conda-forge -n pymc_511 "pymc>=4"
```

Quando vi verrà chiesto se procedere, rispondete di sì:

```bash
Proceed ([y]/n)? y
```

Dobbiamo ora attivare l'ambiente virtuale che abbiamo creato:

```bash
conda activate pymc_511
```

Procediamo con l'installazione di `bambi` ("BAyesian Model-Building Interface"):

```bash
conda install -c conda-forge bambi
```

Aggiungiamo le seguenti librerie:

```bash
conda install -c conda-forge statsmodels
conda install -c anaconda seaborn
conda install -c anaconda graphviz
conda install -c conda-forge watermark
```

Questo conclude la creazione dell'ambiente virtuale `pymc_511`.

#### Altri comandi utili

Un ambiente virtuale è contenuto in una singola cartella. Possiamo ottenere l'indirizzo di questa cartella con:

```bash
conda info -e
```

Ciò rende molto facile rimuovere completamente un ambiente virtuale. Per fare questo, supponendo che l'ambiente virtuale si chiami `my_env`, usiamo:

```bash
conda env remove -n my_env
```

Se vogliamo solo rimuovere una libreria dall'ambiente virtuale (supponiamo si chiami `package_name`), usiamo:

```bash
conda remove package_name
```

## Installazione di un ambiente di sviluppo integrato (IDE)

Dopo avere installato Python e le librerie necessarie, la seconda cosa che dobbiamo fare è installare un ambiente di sviluppo integrato (IDE). Un ambiente di sviluppo integrato è un'applicazione software che fornisce vari strumenti per semplificare la programmazione. 

Ci sono molti IDE diversi. Vi suggerisco di usare Visual Studio Code, un IDE gratuito e open source di Microsoft disponibile per tutti i principali sistemi operativi. Proprio come Python, anche Visual Studio può essere esteso con pacchetti, e sono questi pacchetti, chiamati estensioni, che lo rendono così utile. Visual Studio Code può essere scaricato dalla seguente [pagina web](https://code.visualstudio.com/). 

Visual Studio Code supporta la programmazione sia mediante script che mediante notebook Jupyter. Mentre gli script contengono principalmente codice (e hanno l'estensione file `.py`), i notebook possono contenere testo e codice in diversi blocchi chiamati "celle". I notebook Jupyter hanno l'estensione file `.ipynb`. 

Dopo aver installato e aperto Visual Studio Code, andate alla scheda "estensioni" sulla barra verticale delle icone sul lato sinistro (è l'icona che assomiglia a 4 quadrati). Da lì potete installare l'estensione Python per Visual Studio Code, che potete cercare utilizzando la casella di testo all'interno del pannello delle estensioni di Visual Studio Code.

Per poter utilizzare i notebook Jupyter in Visual Studio Code, è necessario disporre della libreria `ipython`, che al momento risulta essere mancante. Si consiglia di procedere con l'installazione della libreria ipython non appena richiesta dal sistema. In particolare, al momento dell'apertura del primo Jupyter Notebook all'interno di VSCode, l'utente sarà sollecitato ad installare la suddetta libreria, pertanto si raccomanda di fornire una risposta positiva a tale richiesta.
